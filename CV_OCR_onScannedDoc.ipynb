```python
# Import libraries 
from PIL import Image
import pytesseract
import sys
from pdf2image import convert_from_path
import os
import cv2
```

```python
# Path of the scanned pdf  
PDF_file = "sbiform.pdf"
```

## Part -1 
Converting PDF to images  : Each page of the PDF is stored as an image file.

```python
# Store all the pages of the PDF in a variable 
pages = convert_from_path(PDF_file, 500)
#pages = convert_from_path(pdfs, 350,poppler_path=r'C:\Program Files\poppler-0.68.0\bin')
```

```python
counterForImage = 1 # Counter to store images of each page of PDF to image 
```

```python
# Iterate through all the pages stored above 
for page in pages:
    filename = "page_" + str(counterForImage) + ".jpg"
    page.save(filename, 'JPEG')                       # Save the image of the page in system
    counterForImage = counterForImage + 1             # Increment the counter to update filename
```

## Part -2 
Recognizing text from the images using OCR 

```python
# Variable to get count of total number of pages 
filelimit = counterForImage - 1

# Creating a text file to write the output 
outfile = "output.txt"
```

```python
# Open the file in append mode so that All contents of all images are added to the same file 
f = open(outfile, "a")

# Iterate from 1 to total number of pages 
for i in range(1, filelimit + 1):

    filename = "page_" + str(i) + ".jpg"
    pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
    # Recognize the text as string in image using pytesserct
    text = str(((pytesseract.image_to_string(Image.open(filename)))))

    text = text.replace('-\n', '')

    # Write the processed text to the file.
    f.write(text)

# Close the file after writing all the text. 
f.close()
```

```python
f = open(outfile, "r")
print(f.read()) 
```

The below code can be used for marking the regions of interest in the image 

```python
import cv2
import pytesseract

img = cv2.imread('page_1.jpg')

h, w, c = img.shape
boxes = pytesseract.image_to_boxes(img) 
for b in boxes.splitlines():
    b = b.split(' ')
    img = cv2.rectangle(img, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (0, 255, 0), 2)

cv2.imshow('img', img)
cv2.waitKey(0)
```
